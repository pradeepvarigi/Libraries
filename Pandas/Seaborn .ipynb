# Visualizing statistical relationships

Statistical analysis is a process of understanding how variables in a dataset relate to each other and how those relationships depend on other variables. Visualization can be a core component of this process because, when data are visualized properly, the human visual system can see trends and patterns that indicate a relationship.

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#load data already preloaded in seaborn library
tips=sns.load_dataset('tips')
tips.head()
```

```python
#Here i want to see what is the relationship b/w total bill and tip
#relplot:-which is called relational plot it is basically show a relation between the data set 
#data=tips:- which is data set(Here we are calling a data set)
sns.relplot(x='total_bill',y='tip',data=tips)
```

```python
tips['smoker'].value_counts()
```

```python
#hue:-basically used to change the color
#hue='smoker':- if i want to find the person who pays heighest bill and tips is a smoker are not
#style='timer':- here we come to find out that the higest tip paid person came for dinner or Lunch
sns.relplot(x='total_bill',y='tip',data=tips,hue='smoker',style='time')
#so here if you observe the higest paid person came for Dinner because it was in circle
```

```python
tips.head()
```

```python
sns.relplot(x='total_bill',y='tip',style='size',data=tips)
```

```python
#hue is basically plots a color
sns.relplot(x='total_bill',y='tip',hue='size',data=tips)
```

```python
#here iam plotting  all the above thing at the same time (i.e hue, style,size)
sns.relplot(x='total_bill',y='tip',data=tips,size='size',hue='smoker',style='time')
```

```python
tips.head()
```

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
##Catplot():-means categorical plot
#when ever u have some categories of data then you want to perform some plotting operation then u can use this
```

```python
sns.catplot(x='day',y='total_bill',data=tips)
```

```python
#if i wanted to flip above data
sns.catplot(y='day',x='total_bill',data=tips)
```

```python
#jitter=False:- Jitter will show u the data linerly i.e one single line interms of hieght
sns.catplot(x='day',y='total_bill',data=tips,jitter=False)
```

```python
#swarm:- will try to give you flower kind of a plot
sns.catplot(x='day',y='tip',data=tips,hue='size',kind='swarm')
```

```python
tips.head(1)
```

```python
sns.catplot(x='smoker',y='tip',data=tips)
# here i want to see who has give higest tip whether smoker or nosmoker
#Here i was also observed that no smoker have given more tips compered to smoker in no smoker the graph is os dense
```

```python
#BoxPlot:-By default it will try to give you a outliers
#outliers is something which beyond the quartile range
```

```python
sns.catplot(x='smoker',y='tip',data=tips,order=['No','Yes'])
```

```python
sns.catplot(x='day',y='total_bill',kind='box',hue='sex',data=tips)
#max=Q3+1.5*IQR(High outlier)
#min=Q1-1.5*IQR(Low outlier)
#Q1=25th Percentail
#Q2=50th Percentail(i.e median)
#Q3=75th Percentail
```

```python
dia=sns.load_dataset('diamonds')
dia.head()
```

```python
#Here categorical vs numerical
sns.catplot(x='color',y='price',kind='boxen',data=dia)
#Boxen plot is almost simlar to boxen plot but the graph which it us trying to give is little bit different
```

```python
sns.catplot(x='total_bill',y='day',kind='violin',data=tips,split=True,hue='sex')
# we don't use this graph anywhere it just a fancy graph
#it always try to represent a data in terms of viloine
```

```python
#if want plot one garph  inside the another graph
g=sns.catplot(x='day',y='total_bill',kind='violin',inner=None,data=tips)
```

```python
sns.swarmplot(x='day',y='total_bill',color='k',size=3,data=tips)
```

```python
g=sns.catplot(x='day',y='total_bill',kind='violin',inner=None,data=tips)
sns.swarmplot(x='day',y='total_bill',color='k',size=3,data=tips,ax=g.ax)
```
